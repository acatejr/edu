```python
import numpy as np
from numpy.random import randn
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels
%matplotlib inline
```

```python
data1 = randn(100)
```

```python
data2 = randn(100)
```

```python
sns.boxplot(data=[data1, data2])
```

```python
sns.boxplot(data=[data1, data2], whis=np.inf)
```

```python
sns.boxplot(data=[data1,data2], orient='horiz')
```

```python
# Normal dist
data1 = stats.norm(0,5).rvs(100)

# Two gamma dist. Conctatentated
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        stats.gamma(5).rvs(50)])

# Box plot both data1 and data2
sns.boxplot(data=[data1, data2], whis=np.inf)
```

```python
sns.violinplot(data=[data1,data2])
```

```python
sns.violinplot(data=[data1,data2], bw=1)
```

```python
sns.violinplot(data=data1, inner='stick')
```
